```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fft(N):
    return N * np.log2(N)

def dft(N):
    return N *N
```

```python
I = [i for i in range (4, 20)]
N = [2**i for i in I]
```

```python
fft_list = np.asarray( [ fft(n) for n in N] )
dft_list = np.asarray( [ dft(n) for n in N] )
reduction = dft_list/ fft_list
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,5))

ax1.plot(I, dft_list, label="DFT")
ax1.plot(I, fft_list, label="FFT")
ax1.scatter(I, dft_list)
ax1.scatter(I, fft_list)
ax1.set_xlabel(r"i such that $N=2^i$", fontsize=14)
ax1.set_ylabel("O(N)", fontsize=14)
# ax1.set_xscale('log')
# ax1.set_yscale('log')
ax1.set_title(r"O(N) vs i for $N=2^i$", fontsize=16)

ax1.legend(fontsize=14)


ax2.plot(I, reduction, label="DFT/ FFT")
ax2.scatter(I, reduction)

# ax2.plot(N, fft_list, label="FFT")
ax1.set_xlabel(r"i such that $N=2^i$", fontsize=14)
ax2.set_ylabel("Reduction Factor", fontsize=14)
# ax2.set_xscale('log')
# ax2.set_yscale('log')
ax2.set_title(r"Reduction Factor vs i for $N=2^i$", fontsize=16)
ax2.legend()
fig.tight_layout()
```
